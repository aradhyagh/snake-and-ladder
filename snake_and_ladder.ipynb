```python
import random

class SnakeAndLadder:
    def __init__(self):
        self.snakes = {
            99: 78, 95: 75, 92: 88, 89: 68, 74: 53,
            64: 60, 62: 19, 49: 11, 46: 25, 16: 6
        }
        self.ladders = {
            2: 38, 7: 14, 8: 31, 15: 26, 21: 42,
            28: 84, 36: 44, 51: 67, 71: 91, 78: 98, 87: 94
        }

        self.players = {}
        self.num_players = 0

    def setup(self):
        # Number of players
        while True:
            try:
                self.num_players = int(input("Enter number of players (1-4): "))
                if 1 <= self.num_players <= 4:
                    break
                else:
                    print("Please enter between 1 to 4 players.")
            except ValueError:
                print("Invalid input, try again.")

        # Player names
        for i in range(1, self.num_players + 1):
            name = input(f"Enter Player {i} name: ")
            self.players[name] = 0

        print("\nStarting positions:")
        for name, pos in self.players.items():
            print(f"{name}: {pos}")

    def roll_dice(self):
        return random.randint(1, 6)

    def play(self):
        winner = None
        while not winner:
            for player in self.players:
                input(f"\n{player}'s turn. Press Enter to roll the dice...")
                dice = self.roll_dice()
                print(f"{player} rolled a {dice}")

                new_pos = self.players[player] + dice

                if new_pos > 100:
                    print(f"{player} cannot move, needs exact roll to win.")
                else:

                    if new_pos in self.snakes:
                        print(f"Oh no! {player} got bitten by a snake at {new_pos} → {self.snakes[new_pos]}")
                        new_pos = self.snakes[new_pos]
                    elif new_pos in self.ladders:
                        print(f"Yay! {player} climbed a ladder at {new_pos} → {self.ladders[new_pos]}")
                        new_pos = self.ladders[new_pos]

                    self.players[player] = new_pos
                    print(f"{player} moved to {new_pos}")

                    if new_pos == 100:
                        winner = player
                        break


                print("\nCurrent Positions:")
                for name, pos in self.players.items():
                    print(f"{name}: {pos}")

        print(f"\n {winner} wins the game!")


if __name__ == "__main__":
    game = SnakeAndLadder()
    game.setup()
    game.play()
```
